## Source code for generating Fig. 9 in the paper ["STAR-Magic Mutation: Even More Efficient Analog Rotation Gates for Early Fault-Tolerant Quantum Computer"](https://arxiv.org/abs/2603.22891)

```python
import sys
import os
import pathlib
import numpy as np
import matplotlib.pyplot as plt

__file__ = os.path.abspath('')    #`__file__` is not defined on jupyter file (not needed in python file)
src_path = pathlib.Path(__file__).parent / 'src'
assert src_path.exists()
sys.path.append(str(src_path))
from starsim.util.plot_parameters import set_parameter_for_plot
from modules_for_v3 import calc_tradeoff_for_cultivation, calc_tradeoff_for_mutation, calc_tradeoff_for_mutation_without_preparation
```

# Plot tradeoff between speed and accuracy (with limited ancilla patches)

```python
set_parameter_for_plot(6)


# only cultivation
N_grid = 30
epsilon_list = np.logspace(-12, -3, N_grid)
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(epsilon_list):
    p_list[i], t_list[i] = calc_tradeoff_for_cultivation(epsilon)
plt.plot(p_list, t_list, marker="s", label="Cultivation", alpha=0.7)


# STAR-magic mutation
N_grid = 16
d = 21

## The case of theta = 1e-3
theta = 1e-3
ratio_list = [2**i for i in range(N_grid)]
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(ratio_list):
    p_list[i], t_list[i] = calc_tradeoff_for_mutation(d, theta=theta, theta_th=theta*ratio_list[i])
plt.plot(p_list, t_list, marker="o", label=r"Mutation ($\theta_L=10^{-3}$)", alpha=0.7)


## The case of theta = 1e-4
theta = 1e-4
ratio_list = [2**i for i in range(N_grid)]
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(ratio_list):
    p_list[i], t_list[i] = calc_tradeoff_for_mutation(d, theta=theta, theta_th=theta*ratio_list[i])
plt.plot(p_list, t_list, marker="o", label=r"Mutation ($\theta_L=10^{-4}$)", alpha=0.7)

## The case of theta = 1e-5
theta = 1e-5
ratio_list = [2**i for i in range(N_grid)]
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(ratio_list):
    p_list[i], t_list[i] = calc_tradeoff_for_mutation(d, theta=theta, theta_th=theta*ratio_list[i])
plt.plot(p_list, t_list, marker="o", label=r"Mutation ($\theta_L=10^{-5}$)", alpha=0.7)

## The case of theta = 1e-6
theta = 1e-6
ratio_list = [2**i for i in range(N_grid)]
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(ratio_list):
    p_list[i], t_list[i] = calc_tradeoff_for_mutation(d, theta=theta, theta_th=theta*ratio_list[i])
plt.plot(p_list, t_list, marker="o", label=r"Mutation ($\theta_L=10^{-6}$)", alpha=0.7)

## The case of theta = 1e-7
theta = 1e-7
ratio_list = [2**i for i in range(N_grid)]
p_list = np.zeros(N_grid)
t_list = np.zeros(N_grid)
for i, epsilon in enumerate(ratio_list):
    p_list[i], t_list[i] = calc_tradeoff_for_mutation(d, theta=theta, theta_th=theta*ratio_list[i])
plt.plot(p_list, t_list, marker="o", label=r"Mutation ($\theta_L=10^{-7}$)", alpha=0.7)


plt.xscale("log")
plt.yscale("log")
plt.legend(loc='upper left')
plt.ylabel("Execution time [clock]")
plt.xlabel("Logical error rate")
plt.grid(which='major', color='gray', alpha=0.8, lw = 0.8)
plt.grid(which='minor', color='lightgray', alpha=0.8, lw = 0.8)
```
